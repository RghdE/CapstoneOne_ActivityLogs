### 3. Encrypting company name

```python
#Remove the links that have the company name
def Remove_company_name(x):  
    x = str(x) 
    x = x.replace('sukuk','company')
    x = x.replace('manafatech','comtech')   
    return x
```

```python
for col in df:
  df[col]=df[col].apply(lambda x: Remove_company_name(x))
```

```python
for col in df2:
  df2[col]=df2[col].apply(lambda x: Remove_company_name(x))
```
